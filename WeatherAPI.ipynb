```python
# modules
import json
import requests
import matplotlib.pyplot as plt
import time
import pandas as pd 
import numpy as np
import random
###
from datetime import datetime
from citipy import citipy 
```

```python
# OpenWeather API key
owkey  = "6e6b0f450b212799c36cdc62fa58f591"
```

```python
# empty lists
cities = []
countries = []
counter = 0

# use a loop to create a list of cities
while len(cities) < 500 :
    counter +=1
    
    # Randomly generate latitude
    lat_pos = random.randint(0,2)
    base_lat = random.randint(0,91)
    dec_lat = random.random()/100
    lat = base_lat + dec_lat
    if lat_pos == 1:
        lat = lat * -1
    
    # Randomly generate longitude
    lng_pos = random.randint(0,2)
    base_lng = random.randint(0,181)
    dec_lng = random.random()/100
    lng = base_lng + dec_lng
    if lng_pos == 1:
        lng = lng * -1
    
    # Use citipy module to get nearest city from lat and lng
    city = citipy.nearest_city(lat, lng)
    
    # Verify if the random city is in the list, if not, add it to list
    foundCity = False
    for j in range(len(cities)):
        if city.city_name == cities[j]:
            foundCity = True
    if foundCity==False  :   
        cities.append(city.city_name)
        countries.append(city.country_code)
```

```python
# Build query Url
base_url = "http://api.openweathermap.org/data/2.5/weather?"
api_key = owkey
units = "imperial"

query_url = base_url + "appid=" + api_key + "&units=" + units + "&q="
```

```python
search_list = []
city_list = []
country_list = []

city_counter = 0
print("Calling OpenWeather ")
print("*"*30)
for city in cities:
    city_counter += 1
    print(f"City # {city_counter}: {city}")   
    response = requests.get(query_url + city).json()
    city_id = response.get("id")
    if response.get("id"):
        print(f" Data collected for city: {city}  city id: {city_id}")
        search_list.append(response)
        city_list.append(city)
        country_list.append(countries[city_counter-1])
    else:
        print(f" No weather data found for city: {city}")
          
print("*"*30)
print("Complete!")
print(len(search_list))
```

```python
# list comprehension
lat = [data.get("coord").get("lat") for data in search_list]
lng = [data.get("coord").get("lon") for data in search_list]
temp = [data.get("main").get("temp") for data in search_list]
humid = [data.get("main").get("humidity") for data in search_list]
cloud =[data.get("clouds").get("all") for data in search_list]
wind = [data.get("wind").get("speed") for data in search_list]

# Create data frame
w_df = pd.DataFrame({"city":city_list,
                            "country":country_list,
                            "lat":lat,
                            "lng":lng,
                            "temp":temp,
                            "humidity":humid,
                            "cloudiness":cloud,
                            "winds":wind})

# rearrange column order
w_df = w_df[["city","country","lat","lng","temp","humidity","cloudiness","winds"]]
w_df.head(3)
```

```python
date = time.strftime("%m/%d/%Y")
plt.scatter(w_df['lat'],w_df['temp'])
plt.title(f" Max Temperature vs. City Latitude {date}")
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
plt.style.use('ggplot')
plt.savefig("Temperature.png")
plt.show()
```
